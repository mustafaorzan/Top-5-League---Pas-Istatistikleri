```python
pass_cmp = df.loc[df['Total-Cmp'] >= df['Total-Cmp'].mean()].sort_values(
    ['Total-Cmp', 'Total-Cmp%'], ascending=False)[['General-Player','General-Pos','General-Squad','Total-Cmp', 'Total-Cmp%']].head(50)
pass_cmp
```

```python
pass_detail_cmp = df.loc[df['Total-Cmp'] >= df['Total-Cmp'].mean()].sort_values(
    ['Total-Cmp', 'Total-Cmp%'], ascending=False)[['General-Player','General-Pos','General-Squad','Total-Cmp', 'Total-Cmp%',
                                                   'Short-Cmp', 'Short-Cmp%', 'Medium-Cmp', 'Medium-Cmp%', 'Long-Cmp', 'Long-Cmp%',]].head(50)
pass_detail_cmp
```

```python
pass_detail_using_cmp = df.loc[df['Total-Cmp'] >= df['Total-Cmp'].mean()].sort_values(
    ['Total-Cmp', 'Total-Cmp%'], ascending=False)[['General-Player','General-Pos','General-Squad','Total-Cmp', 'Total-Cmp%',
                                                   'Short-Cmp', 'Short-Cmp%', 'Medium-Cmp', 'Medium-Cmp%', 'Long-Cmp', 'Long-Cmp%',]].head(50)
pass_detail_using_cmp['Short-Use'] = np.round( ((pass_detail_using_cmp['Short-Cmp'] / pass_detail_using_cmp['Total-Cmp'])*100), 1)
pass_detail_using_cmp['Medium-Use'] = np.round( ((pass_detail_using_cmp['Medium-Cmp'] / pass_detail_using_cmp['Total-Cmp']*100)), 1)
pass_detail_using_cmp['Long-Use'] = np.round( ((pass_detail_using_cmp['Long-Cmp'] / pass_detail_using_cmp['Total-Cmp'])*100), 1)

pass_detail_using_cmp [['General-Player','General-Pos','General-Squad','Total-Cmp','Short-Cmp','Short-Use', 'Medium-Cmp','Medium-Use','Long-Cmp','Long-Use']]
```

```python
pass_cmp_long = df.loc[df['Long-Cmp'] >= df['Long-Cmp'].mean()].sort_values(
    ['Long-Cmp', 'Long-Cmp%'], ascending=False)[['General-Player','General-Pos','General-Squad','Long-Cmp', 'Long-Cmp%']].head(50)
pass_cmp_long
```

```python
pass_1_3 = df.sort_values(['General-1/3'],ascending=False)[
    ['General-Player','General-Pos','General-Squad','General-Ast',
     'Expected-xA','General-KP','General-1/3','General-PPA', 'General-PrgP']].head(50)

pass_kp = df.sort_values(['General-KP'],ascending=False)[
    ['General-Player','General-Pos','General-Squad','General-Ast',
     'Expected-xA','General-KP','General-1/3','General-PPA', 'General-PrgP']].head(50)

pass_ppa = df.sort_values(['General-PPA'],ascending=False)[
    ['General-Player','General-Pos','General-Squad','General-Ast',
     'Expected-xA','General-KP','General-1/3','General-PPA', 'General-PrgP']].head(50)

pass_prgp = df.sort_values(['General-PrgP'],ascending=False)[
    ['General-Player','General-Pos','General-Squad','General-Ast',
     'Expected-xA','General-KP','General-1/3','General-PPA', 'General-PrgP']].head(50)
```

```python
pass_kp[['General-Player','General-Pos','General-Squad', 'General-KP']].head(10)
```

```python
# Final Third and Pass into Pen Area

merge_1 = pd.merge(left=pass_1_3, right=pass_ppa, how='inner', on=['General-Player','General-Pos','General-Squad',
    'General-Ast','Expected-xA','General-KP','General-1/3','General-PPA', 'General-PrgP'])
merge_1.sort_values('General-Ast',ascending=False)
```

```python
# Progressive Pass and Key Pass
merge_2 = pd.merge(left=pass_kp, right=pass_prgp, how='inner', on=['General-Player','General-Pos','General-Squad',
    'General-Ast','Expected-xA','General-KP','General-1/3','General-PPA', 'General-PrgP'])
merge_2.sort_values('General-Ast',ascending=False)
```

```python
merge_1_2 = pd.merge(left=merge_1, right=merge_2, how='inner', on=['General-Player','General-Pos','General-Squad',
    'General-Ast','Expected-xA','General-KP','General-1/3','General-PPA', 'General-PrgP'])
merge_1_2.sort_values('General-Ast',ascending=False)
```
